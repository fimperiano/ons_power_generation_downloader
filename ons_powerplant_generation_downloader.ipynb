# 📚 Libraries

```python
import requests # For making HTTP requests to download data
import os       # For file and directory operations
```

# 📂 Download ONS data

```python
# Loop over available data files (legacy format)
for year in range (2000, 2021 + 1): # The latest data available in this format is from 2021

    url = f"https://ons-aws-prod-opendata.s3.amazonaws.com/dataset/geracao_usina_2_ho/GERACAO_USINA-2_{year}.parquet"

    # Extract the file name from the URL (e.g. GERACAO_USINA-2_2021.parquet)
    file_name = url.split("/")[-1]

    # Define the file path to save the downloaded file
    file_path = os.path.join(os.getcwd(), # Current working directory
                             "data",      # Target 'data' folder
                             file_name    # File name
                             )

    # Send HTTP (GET) request to download the file
    request = requests.get(url)

    # Check if the request was successful (status code 200)
    if request.status_code == 200:
        # Save the file to the specified path
        # The file content comes from the HTTP response body
        with open (file_path, "wb") as local_file:
            local_file.write(request.content)

        print(f"{file_name} - Downloaded Successfully")


    # Handle other HTTP status codes (e.g., 403 Forbidden, 500 Internal Server Error)
    else:
        print(
            f"File download error. Stauts Code: {request.status_code}"
            )
```
