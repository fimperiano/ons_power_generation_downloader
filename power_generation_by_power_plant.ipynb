# 📚 Libraries

```python
import requests # HTTP requests and web services
import os       # folder and files manipulation
```

# 📂 Download ONS data

```python
# Creating the loop for the avilable data (old style data files)
for year in range (2000, 2021 + 1): # The latest data avilable in this old format is for 2021.

    url = f"https://ons-aws-prod-opendata.s3.amazonaws.com/dataset/geracao_usina_2_ho/GERACAO_USINA-2_{year}.parquet"

    # To format the dowloaded files names (e.g. GERACAO_USINA-2_2021.parquet)
    file_name = url.split("/")[-1]

    # Set file the path to save the files
    file_path = os.path.join(os.getcwd(), "data", file_name)

    # Faz a requisição HTTP (GET) para baixar o arquivo
    request = requests.get(url)

    # Successful HTTP request and download
    # STATUS CODE = 200
    if request.status_code == 200:
        # Salvar o arquivo no caminho desejado
        # File = HTTP request conteúdo da resposta da requisição HTTP
        with open (file_path, "wb") as local_file:
            local_file.write(request.content)

        print(f"{file_name} - Downloaded Successfully")


    # Other status code errors
    # STATUS CODE = 403 (Forbidden) / 500 (Internal Server Error)
    else:
        print(
            f"File download error. Stauts Code: {request.status_code}"
            )
```
